```python
for col in follow.columns:
    if 'Unnamed' in col:
        del follow[col]
```

```python
follow['Date_of_dep']=pd.to_datetime(follow['Date_of_dep'])
```

```python
follow['Date_of_visit']=pd.to_datetime(follow['Date_of_visit'])
```

```python
follow.isnull().sum()
```

```python
follow.columns.str.replace(' ','@')
```

```python
def myfillna(series):
    if series.dtype is pd.np.dtype(float):
        return series.fillna('NA')
    elif series.dtype is pd.np.dtype(int):
        return series.fillna('NA')
    else:
        return series.fillna('NA')
```

```python
follow=follow.apply(myfillna)
```
